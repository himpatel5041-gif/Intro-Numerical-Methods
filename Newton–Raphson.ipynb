```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Function and its derivative
def f(x):
    return x**3 - x - 2

def df(x):
    return 3*x**2 - 1

# Newton-Raphson parameters
x0 = 1.5          # initial guess
tol = 1e-10
max_iter = 20

x = x0
values = [x0]

# Newton-Raphson loop
for i in range(max_iter):
    x_new = x - f(x)/df(x)
    values.append(x_new)

    if abs(x_new - x) < tol:
        break
    x = x_new

# Print iterations
print("Iterations:")
for i, val in enumerate(values):
    print(f"Iter {i}: x = {val}")

print("\nFinal root ≈", values[-1])

# Plot convergence
iterations = range(len(values))

plt.figure(figsize=(6,4))
plt.plot(iterations, values)
plt.xlabel("Iteration")
plt.ylabel("Root estimate (x)")
plt.title("Newton-Raphson Convergence")
plt.grid(True)
plt.show()
```
